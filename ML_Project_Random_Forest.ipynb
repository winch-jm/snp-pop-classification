```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
import pandas as pd

# all packages
```

```python
# load the data in/ in this case, I am using a smaller dataset
data = {
    'age':            [22, 45, 31, 60, 28, 52, 39, 24, 47, 33, 55, 41, 26, 63, 37],
    'tenure_months':  [3,  24, 12, 36, 6,  48, 18, 2,  30, 9,  42, 15, 5,  60, 20],
    'monthly_spend':  [30, 90, 55, 120, 40, 100, 75, 25, 95, 50, 110, 80, 35, 130, 70],
    'support_calls':  [5,  1,  3,  0,  4,  1,  2,  6,  1,  3,  0,  2,  5,  0,  2],
    'churned':        [1,  0,  0,  0,  1,  0,  0,  1,  0,  1,  0,  0,  1,  0,  0]
}
df = pd.DataFrame(data)

X = df.drop("churned",axis=1)
Y = df['churned']

#Training/testdataset split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

# creating a random forest object
#42 is the seed for the randomness
# should we change the number of trees? Will create graphs and see 
RFO = RandomForestClassifier(n_estimators=100, random_state=42)

#training the model
RFO.fit(X_train,Y_train)


#Evaluating the model
pred = RFO.predict(X_test)
print(f"Accuracy: {accuracy_score(Y_test, pred):.2f}")
print(classification_report(Y_test, pred, target_names=["Stayed", "Churned"]))

# Feature importance
print("\nFeature Importances:")
for feat, imp in sorted(zip(X.columns, RFO.feature_importances_), key=lambda x: -x[1]):
    print(f"  {feat}: {imp:.4f}")

# Predict on a new customer
new_customer = pd.DataFrame([{
    'age': 27,
    'tenure_months': 4,
    'monthly_spend': 35,
    'support_calls': 5
}])
prob = RFO.predict_proba(new_customer)[0][1]
print(f"\nNew customer churn probability: {prob:.0%}")





```
